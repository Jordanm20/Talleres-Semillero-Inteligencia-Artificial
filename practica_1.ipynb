```python
from huggingface_hub import login
from transformers import pipeline

token = ""
login(token)

modelo = "datificate/gpt2-small-spanish"

modelo_chistes = pipeline(
    "text-generation",
    model=modelo,
    device=-1
)

print("Modelo cargado correctamente:", modelo)
```

```python
prompt = "Cuéntame un chiste corto sobre inteligencia artificial."

resultado = modelo_chistes(
    prompt,
    max_new_tokens=80,
    temperature=0.9,
    top_p=0.95,
    do_sample=True,
    return_full_text=False
)[0]["generated_text"]

print("🎭 Prompt:")
print(prompt)

print("\n🤣 Chiste generado:")
print(resultado)
```

```python
prompt_simple = "Cuéntame un chiste corto sobre inteligencia artificial."

resultado_simple = modelo_chistes(
    prompt_simple,
    max_new_tokens=80,
    temperature=0.9,
    top_p=0.95,
    do_sample=True
)[0]["generated_text"]

print("PROMPT SIMPLE:")
print(prompt_simple)

print("\nCHISTE GENERADO:")
print(resultado_simple)
```

```python
prompt_contexto = """
Escribe un chiste corto sobre inteligencia artificial.
El chiste debe ser fácil de entender para estudiantes universitarios.
Debe ser gracioso, breve y tener un remate final.

Chiste:
"""

resultado_contexto = modelo_chistes(
    prompt_contexto,
    max_new_tokens=80,
    temperature=0.9,
    top_p=0.95,
    do_sample=True,
    repetition_penalty=1.15,
    return_full_text=False
)[0]["generated_text"]

print("PROMPT CON CONTEXTO:")
print(prompt_contexto)

print("\nCHISTE GENERADO:")
print(resultado_contexto)
```

```python
prompt_mejorado = """
Chiste corto sobre inteligencia artificial:

Pregunta: ¿Por qué la inteligencia artificial no fue a clases?
Respuesta:
"""

resultado_mejorado = modelo_chistes(
    prompt_mejorado,
    max_new_tokens=50,
    temperature=0.8,
    top_p=0.9,
    do_sample=True,
    repetition_penalty=1.25,
    return_full_text=False
)[0]["generated_text"]

print("PROMPT MEJORADO:")
print(prompt_mejorado)

print("\nCHISTE GENERADO:")
print(resultado_mejorado)
```

```python
configuraciones = [
    {
        "nombre": "Conservador",
        "temperature": 0.3,
        "top_p": 0.80,
        "max_new_tokens": 40
    },
    {
        "nombre": "Balanceado",
        "temperature": 0.7,
        "top_p": 0.90,
        "max_new_tokens": 50
    },
    {
        "nombre": "Creativo",
        "temperature": 0.9,
        "top_p": 0.95,
        "max_new_tokens": 60
    },
    {
        "nombre": "Muy creativo",
        "temperature": 1.2,
        "top_p": 0.98,
        "max_new_tokens": 70
    }
]

resultados_parametros = []

for config in configuraciones:
    resultado = modelo_chistes(
        prompt_mejorado,
        max_new_tokens=config["max_new_tokens"],
        temperature=config["temperature"],
        top_p=config["top_p"],
        do_sample=True,
        repetition_penalty=1.25,
        return_full_text=False
    )[0]["generated_text"]

    resultados_parametros.append({
        "Configuración": config["nombre"],
        "Temperature": config["temperature"],
        "Top_p": config["top_p"],
        "Max_new_tokens": config["max_new_tokens"],
        "Resultado": resultado
    })

    print("=" * 80)
    print("CONFIGURACIÓN:", config["nombre"])
    print("Temperature:", config["temperature"])
    print("Top_p:", config["top_p"])
    print("Max_new_tokens:", config["max_new_tokens"])
    print("\nCHISTE GENERADO:")
    print(resultado)
```

## Análisis de parámetros

Durante la práctica se probaron diferentes valores de `temperature`, `top_p` y `max_new_tokens` para observar cómo afectan la generación del chiste.

- Con `temperature = 0.3`, el modelo fue más conservador. El resultado fue más controlado, pero menos creativo.
- Con `temperature = 0.7`, el modelo generó un resultado más equilibrado entre coherencia y creatividad.
- Con `temperature = 0.9`, el modelo produjo respuestas más variadas y con mayor intención humorística.
- Con `temperature = 1.2`, el modelo fue más creativo, pero también aumentó el riesgo de generar respuestas incoherentes o demasiado largas.

El parámetro `top_p` permitió controlar la diversidad de palabras usadas por el modelo. Valores bajos como `0.80` limitaron más la respuesta, mientras que valores altos como `0.95` o `0.98` permitieron más variedad.

El parámetro `max_new_tokens` controló la longitud máxima del texto generado. Cuando el valor fue bajo, el chiste podía quedar incompleto. Cuando fue alto, el modelo podía extenderse demasiado.

## Mejor prompt utilizado

El mejor prompt utilizado fue:

"Chiste corto sobre inteligencia artificial:

Pregunta: ¿Por qué la inteligencia artificial no fue a clases?
Respuesta:"

## Justificación

Este prompt funcionó mejor porque usa una estructura simple de pregunta y respuesta. En comparación con el prompt largo, el modelo pudo continuar el texto de forma más directa y generar una respuesta más cercana a un chiste. Los modelos GPT-2 pequeños no siempre siguen instrucciones extensas, por eso un formato breve y guiado produjo mejores resultados.

## Conclusión

En esta parte de la práctica se comprobó que la calidad del resultado depende mucho de la forma en que se redacta el prompt. Un prompt simple genera respuestas básicas, mientras que un prompt con rol, contexto y reglas mejora la coherencia del texto generado.

También se observó que los parámetros de generación influyen directamente en la creatividad del modelo. `temperature` controla el nivel de creatividad, `top_p` regula la diversidad de palabras y `max_new_tokens` define la extensión máxima de la respuesta. La mejor configuración fue la que mantuvo un equilibrio entre creatividad y coherencia.
